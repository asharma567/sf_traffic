```python
import pandas as pd
import numpy as np
import matplotlib as plt
import os, sys
%pylab inline
```

```python
bus = pd.read_csv('bus_data/test2.csv')
bus.head()
```

```python
bus = bus[bus["LONGITUDE"] <= -122.335] # remove faulty data
bus = bus[bus["LONGITUDE"] >= -122.541]
bus = bus[bus["LATITUDE"] <= 37.845]
bus = bus[bus["LATITUDE"] >= 37.68]
bus.describe()
```

```python
uber = pd.read_csv('uber_data/all.tsv',sep='\t')
uber.head()
```

```python
uber = uber[uber["x"] <= -120] # remove faulty data
uber = uber[uber["x"] >= -125]
uber = uber[uber["y"] <= 40]
uber = uber[uber["y"] >= 35]
uber.describe()
```
