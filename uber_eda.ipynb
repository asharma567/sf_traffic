```python
import pandas as pd
import numpy as np
import matplotlib as plt
%pylab inline
```

```python
df = pd.read_csv('uber_data/all.tsv',sep='\t')

#def date2int(x):
#    m = x[5:7]
#    d = x[8:10]
#    if m == '05':
#        return int(d)-17
#    elif m == '06':
#        return int(d)+14
#    elif m == '07':
#        return int(d)+34

df["year"] = [int(t[0:4]) for t in df["datetime"]]
df["month"] = [int(t[5:7]) for t in df["datetime"]]
df["day"] = [int(t[8:10]) for t in df["datetime"]]
df["hr"] = [int(t[11:13]) for t in df["datetime"]]
df["min"] = [int(t[14:16]) for t in df["datetime"]]
df["sec"] = [int(t[17:19]) for t in df["datetime"]]
df["time"] = 100*df["sec"]/60 + 10000*df["min"]/60 + df["hr"]*10000
df.head()
```

```python
df = df[df["x"] <= -120] # remove faulty data
df = df[df["x"] >= -125]
df = df[df["y"] <= 40]
df = df[df["y"] >= 35]
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,15))
ax = plt.subplot(111,axisbg='black')
#color = df['time']/df['time'].max()
plt.scatter(df['x'],df['y'],linewidth=0,s=1,c='orange',alpha=0.1)
plt.xticks([])
plt.yticks([])
```

```python
plt.figure(figsize=(20,20))

daydict = {}
daydict[1] = 'Monday'
daydict[2] = 'Tuesday'
daydict[3] = 'Wednesday'
daydict[4] = 'Thursday'
daydict[5] = 'Friday'
daydict[6] = 'Saturday'
daydict[7] = 'Sunday'

for i in xrange(7):
    x = df[df['day'] == i+1]['x']
    y = df[df['day'] == i+1]['y']
    c = df[df['day'] == i+1]['ride']
    ax = plt.subplot(3,3,i+1,axisbg='black')
    plt.scatter(x,y,linewidth=0,s=10,c=c,alpha=0.3)
    plt.xlim(-122.6, -122.15)
    plt.ylim(37.5, 38.0)
    plt.xticks([])
    plt.yticks([])
    plt.title(daydict[i+1])
```
