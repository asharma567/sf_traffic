```python
import pandas as pd
import numpy as np
import matplotlib as plt
%pylab inline
```

```python
df = pd.read_csv('uber_data/all.tsv',sep='\t')

#def date2int(x):
#    m = x[5:7]
#    d = x[8:10]
#    if m == '05':
#        return int(d)-17
#    elif m == '06':
#        return int(d)+14
#    elif m == '07':
#        return int(d)+34

df["year"] = [int(t[0:4]) for t in df["datetime"]]
df["month"] = [int(t[5:7]) for t in df["datetime"]]
df["day"] = [int(t[8:10]) for t in df["datetime"]]
df["hr"] = [int(t[11:13]) for t in df["datetime"]]
df["min"] = [int(t[14:16]) for t in df["datetime"]]
df["sec"] = [int(t[17:19]) for t in df["datetime"]]
df["time"] = 100*df["sec"]/60 + 10000*df["min"]/60 + df["hr"]*10000
df.head()
```

```python
# some rides have duplicated datapoints
df.drop_duplicates(inplace=True)

# remove remaining rides that are shorter than 5 minutes
df = df[df.groupby(['ride'])['ride'].transform('count') >= 75] 
# should actually do this by filtering trip duration, rather than value counts
```

```python
df = df[df["x"] <= -120] # remove faulty data
df = df[df["x"] >= -125]
df = df[df["y"] <= 40]
df = df[df["y"] >= 35]
```

```python
# Central finite difference function to estimate derivatives
def deriv(t,x):
    "Finite difference derivative of the function f"
    n = len(x)
    d = zeros(n,'float') # assume float
    # Use centered differences for the interior points, one-sided differences for the ends
    for i in range(1,n-1):
        d[i] = (x.iloc[i+1]-x.iloc[i])/(t.iloc[i+1]-t.iloc[i])
    d[0] = (x.iloc[1]-x.iloc[0])/(t.iloc[1]-t.iloc[0])
    d[n-1] = (x.iloc[n-1]-x.iloc[n-2])/(t.iloc[n-1]-t.iloc[n-2])
    return d
```

```python
grouped = df.groupby(['ride'])
count = 0
for name, group in grouped:
    speedx = deriv(group['time'], group['x'])
    speedy = deriv(group['time'], group['y'])
    if count == 0:
        speeddf = pd.DataFrame(np.sqrt(speedx**2 + speedy**2),index=[group.index],columns=['speed'])
    else:
        speeddf = pd.concat([speeddf,pd.DataFrame(np.sqrt(speedx**2 + speedy**2),index=[group.index],columns=['speed'])])
    count += 1

df = df.join(speeddf)
df.head()
```

```python
count = 0
for name, group in df.groupby(['ride']):
    accel = deriv(group['time'], group['speed'])
    if count == 0:
        acceldf = pd.DataFrame(accel,index=[group.index],columns=['accel'])
    else:
        acceldf = pd.concat([acceldf,pd.DataFrame(accel,index=[group.index],columns=['accel'])])
    count += 1
df = df.join(acceldf)
df.head()
```
